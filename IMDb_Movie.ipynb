# Data Analysis

### Let's Reduce the digits
 

The numbers in the `budget` and `gross` are too big, compromising its readability. Let's convert the unit of the `budget` and `gross` columns from `$` to `million $` first.

```python
# Divide the 'gross' and 'budget' columns by 1000000 to convert '$' to 'million $'

movies['Gross'] = movies['Gross']/1000000
movies['budget'] = movies['budget']/1000000
```

```python
#to check if gross and budget are converted to million
movies[['Gross','budget']].head()
```

### Let's Talk Profit!

Calculating profit for all the movies and inserting it in new column

```python
# Creating new column 'profit' by subtracting the 'budget' column from the 'gross' column

movies['Profit'] = movies['Gross'] - movies['budget']
```

```python
# to verify 'profit' column is created with correct values
movies[['Gross','budget','Profit']].head()
```

```python
# Sorting the dataframe with the 'profit' column as reference using the 'sort_values' function.

sorted_movies = movies.sort_values('Profit', ascending = False)
```

```python
# Geting top 10 profitable movies by using position based indexing.

top10 = sorted_movies.iloc[0:10,:]
top10
```

```python
def bar_plot_fun(frame, col1, col2):
    plt.figure(figsize = (10,7))

    chart = sns.barplot(x = frame[col1], y = frame[col2])
    chart.set_xticklabels(chart.get_xticklabels(), fontsize=14, rotation=45, ha='right')
    for p in chart.patches:
        chart.annotate(format(p.get_height(), '.1f'), 
                           (p.get_x() + p.get_width() / 2, p.get_height()), 
                           ha = 'center', va = 'center', 
                           xytext = (5, 10), 
                           textcoords = 'offset points',fontsize=12)
    plt.title('Chart for '+col1+' Vs '+col2, fontsize = 25)
    plt.xlabel(col1, fontsize = 20)
    plt.ylabel(col2, fontsize = 20)
    plt.show()
```

```python
bar_plot_fun(top10, "Title", "Profit")
```

#### Star Wars: The Force Awakens has made the most profit from Top 100 movies with 691.7 million dollars. The Avengers and Deadpool holds second and third position with 403 and 305 millions dollars respectively.

### The General Audience and the Critics

Column `MetaCritic` is a very popular website where an average score is determined through the scores given by the top-rated critics. Also there is another column `IMDb_rating` which gives the IMDb rating of a movie. This rating is determined by taking the average of hundred-thousands of ratings from the general audience. 

#### Finding out the highest rated movies which have been liked by critics and audiences alike.

```python
# Changing the scale of MetaCritic

movies['MetaCritic'] = movies.MetaCritic/10
movies.MetaCritic
```

```python
# Finding average ratings

movies['Avg_rating'] = (movies['MetaCritic'] + movies['IMDb_rating'])/2
movies.head()
```

```python
#Sorting in descending order of average rating

movies.sort_values('Avg_rating', ascending = False, inplace = True)
movies.head()
```

```python
# Find the movies with metacritic-rating < 0.5 and also with the average rating of >8

movies['Diff_rating'] = abs(movies['IMDb_rating'] - movies['MetaCritic'])
UniversalAcclaim = movies[(movies['Diff_rating']<0.5) & (movies['Avg_rating'] >= 8)]
UniversalAcclaim
```

```python
bar_plot_fun(UniversalAcclaim, "Title", "Avg_rating")
```

#### We got above top rated list of movies by finding average of Critics rating and IMDB rating. These movies were liked by general audience and critics both.

## Know your Audience!

If we take a look at the last columns in the dataframe, most of these are related to demographics of the voters. We also have three genre columns indicating the genres of a particular movie. 

There are 3 columns in the dataframe - `genre_1`, `genre_2`, and `genre_3`. We need to aggregate a few values over these 3 columns for further analysis.

```python
# to fetch genre columns and columns consisting votes

genre = movies.iloc[:,11:14]
votes = movies.iloc[:,16:60]
```

```python
# Create the dataframe df_by_genre

df_by_genre = pd.concat([genre, votes], axis = 1)
df_by_genre
```

```python
# to verify number of columns (should be 47)
df_by_genre.shape
```

```python
# Create a column cnt and initialize it to 1

df_by_genre['cnt'] = 1
```

```python
# to check new column
df_by_genre.head()
```

```python
# Group the movies by individual genres

df_by_g1 = df_by_genre.groupby(['genre_1'], axis = 0).sum()
df_by_g1
```

```python
df_by_g2 = df_by_genre.groupby(['genre_2'], axis = 0).sum()
df_by_g2
```

```python
df_by_g3 = df_by_genre.groupby(['genre_3'], axis = 0).sum()
df_by_g3
```

```python
# Add the grouped data frames and store it in a new data frame

df_add = df_by_g1.add(df_by_g2, fill_value = 0).add(df_by_g3, fill_value = 0)
df_add
```

```python
# Extract genres with atleast 10 occurences

genre_top10 = df_add[df_add['cnt'] >= 10]
genre_top10 
```

```python
# verify rows of genre_top10

genre_top10.shape
```

```python
# Take the mean for every column by dividing with cnt 

genre_top10.iloc[:,0:44] = genre_top10.iloc[:,0:44].div(genre_top10['cnt'], axis=0)
genre_top10.iloc[:,0:44]
```

```python
# Rounding off the columns of Votes to two decimals

genre_top10.loc[:, "VotesM":"VotesnUS"] = genre_top10.loc[:, "VotesM":"VotesnUS"].round(decimals = 2)

# Checking the dataframe
genre_top10
```

```python
# Converting CVotes to int type

genre_top10.loc[:, "CVotes10":"CVotesnUS"] = genre_top10.loc[:, "CVotes10":"CVotesnUS"].astype("int32")

#checking dataframe

genre_top10
```

## Gender and Genre

The Votes- and CVotes-related columns, you might have noticed the suffixes `F` and `M` indicating Female and Male. Since we have the vote counts for both males and females, across various age groups, let's now see how the popularity of genres vary between the two genders in the dataframe. 


```python
# 1st set of heat maps for CVotes-related columns

plt.figure(figsize = (20,10))

# Subplot_1: Male Votes across genres 
plt.subplot(1,2,1)
sns.heatmap((pd.pivot_table(data = genre_top10, index = genre_top10.index, values = ["CVotesU18M","CVotes1829M","CVotes3044M","CVotes45AM"], aggfunc = "mean" )), annot = True, cmap = "YlGnBu")
plt.title("Male Votes for various genres", fontdict = {"fontsize":25, "color": "Blue"})

# Subplot_2: Female Votes across genres 
plt.subplot(1,2,2)
sns.heatmap((pd.pivot_table(data = genre_top10, index = genre_top10.index, values = ["CVotesU18F","CVotes1829F","CVotes3044F","CVotes45AF"], aggfunc = "mean" )), annot = True, cmap = "YlGnBu")
plt.title("Female Votes for various genres", fontdict = {"fontsize":25, "color": "Blue"})

plt.show()
```

#### A few inferences that can be seen from the heatmap above is that males have voted more than females, and Sci-Fi appears to be most popular among the 18-29 age group irrespective of their gender.
#### Under 18 age group is least voted group. Also Age category of 18-29 have voted more than other categories.
#### Romance and comedy genre is more popular in females as compared to males.
#### Action categorty is more popular in males as compared to females irrespective of age categories
